## **í…ìŠ¤íŠ¸ ë¶„í•  (Splitting)**  

ì •ë³´ ê²€ìƒ‰ ë° í›„ì† ì§ˆë¬¸-ì‘ë‹µ(question-answering) ì‘ì—…ì„ í•˜ëŠ”ë° **í˜ì´ì§€ ë‹¨ìœ„** ì‚¬ìš©ì€ ë„ˆë¬´ ê±°ì¹œ í‘œí˜„ì¼ ìˆ˜ ìˆìŠµë‹ˆë‹¤. ìš°ë¦¬ì˜ ìµœì¢… ëª©í‘œëŠ” **ì…ë ¥ ì¿¼ë¦¬ì— ë‹µë³€í•  ìˆ˜ ìˆëŠ” `Document` ê°ì²´ë¥¼ ê²€ìƒ‰í•˜ëŠ” ê²ƒ**ì´ë¯€ë¡œ, PDFë¥¼ ë” ì„¸ë¶„í™”í•˜ë©´ ë¬¸ì„œì˜ í•´ë‹¹ë˜ëŠ” ë¶€ë¶„ì´ ì£¼ë³€ í…ìŠ¤íŠ¸ì— ì˜í•´ ì˜ë¯¸ê°€ íë ¤ì§€ëŠ” ê²ƒì„ ë°©ì§€í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.  

ì´ë¥¼ ìœ„í•´ í…ìŠ¤íŠ¸ ë¶„í• ê¸°(text splitters)ë¥¼ ì‚¬ìš©í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤. ì—¬ê¸°ì„œëŠ” **ë¬¸ìë¥¼ ê¸°ì¤€ìœ¼ë¡œ ë¶„í• í•˜ëŠ” ê°„ë‹¨í•œ í…ìŠ¤íŠ¸ ë¶„í• ê¸°**ë¥¼ ì‚¬ìš©í•  ê²ƒì…ë‹ˆë‹¤.  

- ë¬¸ì„œë¥¼ **1000ì ë‹¨ìœ„ë¡œ ë¶„í• **í•©ë‹ˆë‹¤.  
- ê° ì²­í¬(chunk) ì‚¬ì´ì—ëŠ” **200ìì˜ ì¤‘ì²©(overlap)** ì´ ìˆìŠµë‹ˆë‹¤.  

**ì¤‘ì²©**ì€ ì¤‘ìš”í•œ ë¬¸ë§¥(context)ê³¼ ë¬¸ì¥(statement)ì´ ë¶„ë¦¬ë˜ëŠ” ê²ƒì„ ì™„í™”í•˜ëŠ” ë° ë„ì›€ì´ ë©ë‹ˆë‹¤.  

ìš°ë¦¬ëŠ” **RecursiveCharacterTextSplitter**ë¥¼ ì‚¬ìš©í•  ê²ƒì…ë‹ˆë‹¤. ì´ ë¶„í• ê¸°ëŠ” ì¼ë°˜ì ì¸ êµ¬ë¶„ì(ì˜ˆ: ì¤„ë°”ê¿ˆ)ë¥¼ ì‚¬ìš©í•˜ì—¬ **ì¬ê·€ì ìœ¼ë¡œ ë¬¸ì„œë¥¼ ë¶„í• **í•˜ë©°, ê° ì²­í¬ê°€ ì ì ˆí•œ í¬ê¸°ê°€ ë  ë•Œê¹Œì§€ ë°˜ë³µí•©ë‹ˆë‹¤. **ì¼ë°˜ í…ìŠ¤íŠ¸ ì‚¬ìš© ì‚¬ë¡€ì—ì„œ ê¶Œì¥ë˜ëŠ” í…ìŠ¤íŠ¸ ë¶„í• ê¸°ì…ë‹ˆë‹¤.**  

### **ì¶”ê°€ ì„¤ì •**  
- `add_start_index=True`ë¡œ ì„¤ì •í•˜ë©´ **ê° ë¶„í• ëœ `Document`ê°€ ì›ë³¸ `Document` ë‚´ì—ì„œ ì‹œì‘í•˜ëŠ” ë¬¸ì ì¸ë±ìŠ¤**ê°€ **ë©”íƒ€ë°ì´í„° ì†ì„±(`start_index`)** ìœ¼ë¡œ ë³´ì¡´ë©ë‹ˆë‹¤.  

````python
# RecursiveCharacterTextSplitterë¥¼ ê°€ì ¸ì˜µë‹ˆë‹¤.
# ì´ ë„êµ¬ëŠ” í…ìŠ¤íŠ¸ë¥¼ ì¬ê·€ì ìœ¼ë¡œ ë¶„í• í•˜ë©°, ì¼ë°˜ì ì¸ êµ¬ë¶„ì(ì˜ˆ: ì¤„ë°”ê¿ˆ)ë¥¼ ì‚¬ìš©í•©ë‹ˆë‹¤.
from langchain_text_splitters import RecursiveCharacterTextSplitter

# RecursiveCharacterTextSplitter ì„¤ì •
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,              # ê° ì²­í¬(chunk)ì˜ ìµœëŒ€ ë¬¸ì ìˆ˜
    chunk_overlap=200,           # ì²­í¬ ê°„ ì¤‘ì²©(overlap) ë¬¸ì ìˆ˜
    add_start_index=True        # ê° ì²­í¬ì˜ ì‹œì‘ ì¸ë±ìŠ¤ë¥¼ ë©”íƒ€ë°ì´í„°ë¡œ ì¶”ê°€
)

# ë¬¸ì„œë¥¼ ë¶„í• í•©ë‹ˆë‹¤. 'docs'ëŠ” ë¶„í• í•  ì›ë³¸ ë¬¸ì„œ ëª©ë¡ì…ë‹ˆë‹¤.
all_splits = text_splitter.split_documents(docs)

# ë¶„í• ëœ ë¬¸ì„œì˜ ì´ ê°œìˆ˜
len(all_splits)
````

## **ì„ë² ë”© (Embeddings)**  

**ë²¡í„° ê²€ìƒ‰(Vector Search)** ì€ ë¹„ì •í˜• ë°ì´í„°(ì˜ˆ: ë¹„ì •í˜• í…ìŠ¤íŠ¸)ë¥¼ ì €ì¥í•˜ê³  ê²€ìƒ‰í•˜ëŠ” ì¼ë°˜ì ì¸ ë°©ë²•ì…ë‹ˆë‹¤. í•µì‹¬ ì•„ì´ë””ì–´ëŠ” í…ìŠ¤íŠ¸ì™€ ì—°ê²°ëœ **ìˆ«ì ë²¡í„°(Numeric Vectors)** ë¥¼ ì €ì¥í•˜ëŠ” ê²ƒì…ë‹ˆë‹¤.  

ì¿¼ë¦¬(Query)ê°€ ì£¼ì–´ì§€ë©´, ì´ë¥¼ **ë™ì¼í•œ ì°¨ì›ì˜ ë²¡í„°ë¡œ ì„ë² ë”©(Embedding)** í•˜ê³  **ë²¡í„° ìœ ì‚¬ë„ ì¸¡ì •(metric)** (ì˜ˆ: **ì½”ì‚¬ì¸ ìœ ì‚¬ë„**)ì„ ì‚¬ìš©í•˜ì—¬ ê´€ë ¨ í…ìŠ¤íŠ¸ë¥¼ ì‹ë³„í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.   

### **ëª¨ë¸ ì„ íƒí•˜ê¸°**  

ì›í•˜ëŠ” ì„ë² ë”© ëª¨ë¸ì„ ì„ íƒí•´ ì‚¬ìš©í•©ë‹ˆë‹¤.  
ì˜ˆë¥¼ ë“¤ì–´, OpenAI, Hugging Face, Cohere ë“± ë‹¤ì–‘í•œ ì œê³µìë¥¼ ì‚¬ìš©í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.  

**ì„ë² ë”© ëª¨ë¸ì„ ì„ íƒí•œ í›„ ë²¡í„° ìŠ¤í† ì–´(Vector Store)ì— í…ìŠ¤íŠ¸ë¥¼ ì €ì¥í•˜ê³  ê²€ìƒ‰**í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

````python
from langchain_openai import OpenAIEmbeddings

embeddings = OpenAIEmbeddings(model="text-embedding-3-small")
embeddings
````

````python
# ì¿¼ë¦¬ í…ìŠ¤íŠ¸ë¥¼ ë²¡í„°ë¡œ ì„ë² ë”©í•©ë‹ˆë‹¤.
# ê° ì²­í¬ì˜ ë‚´ìš©ì„ ë²¡í„°ë¡œ ë³€í™˜í•˜ì—¬ ë¹„êµí•©ë‹ˆë‹¤.

# ì²« ë²ˆì§¸ ì²­í¬(all_splits[0])ì˜ ë‚´ìš©ì„ ë²¡í„°ë¡œ ì„ë² ë”©í•©ë‹ˆë‹¤.
vector_1 = embeddings.embed_query(all_splits[0].page_content)

# ë‘ ë²ˆì§¸ ì²­í¬(all_splits[1])ì˜ ë‚´ìš©ì„ ë²¡í„°ë¡œ ì„ë² ë”©í•©ë‹ˆë‹¤.
vector_2 = embeddings.embed_query(all_splits[1].page_content)

# ë‘ ë²¡í„°ì˜ ê¸¸ì´ê°€ ë™ì¼í•œì§€ í™•ì¸í•©ë‹ˆë‹¤.
assert len(vector_1) == len(vector_2), "ë²¡í„° ê¸¸ì´ê°€ ì¼ì¹˜í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤!"

# ë²¡í„°ì˜ ê¸¸ì´ë¥¼ ì¶œë ¥í•©ë‹ˆë‹¤.
print(f"ìƒì„±ëœ ë²¡í„°ì˜ ê¸¸ì´: {len(vector_1)}\n")

# ì²« ë²ˆì§¸ ë²¡í„°ì˜ ì¼ë¶€(ì²˜ìŒ 10ê°œ ìš”ì†Œ)ë¥¼ ì¶œë ¥í•©ë‹ˆë‹¤.
print(vector_1[:10])
````

## **ë²¡í„° ìŠ¤í† ì–´(Vector Stores)**

## **VectorStore ì¿¼ë¦¬í•˜ê¸°**  

**`VectorStore`** ê°€ ë¬¸ì„œë¥¼ í¬í•¨í•˜ê³  ìˆëŠ” ìƒíƒœì—ì„œ, ìš°ë¦¬ëŠ” ì´ë¥¼ ì¿¼ë¦¬í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

**ì„ë² ë”©(Embeddings)** ì€ ì¼ë°˜ì ìœ¼ë¡œ í…ìŠ¤íŠ¸ë¥¼ **'ë°€ì§‘ ë²¡í„°(Dense Vector)'** ë¡œ í‘œí˜„í•©ë‹ˆë‹¤. ì´ë¡œ ì¸í•´ **ì˜ë¯¸ê°€ ìœ ì‚¬í•œ í…ìŠ¤íŠ¸ëŠ” ê¸°í•˜í•™ì ìœ¼ë¡œ ê°€ê¹Œìš´ ìœ„ì¹˜**ì— ìˆê²Œ ë©ë‹ˆë‹¤.

## **ê²€ìƒ‰ê¸° (Retrievers)**  

LangChainì˜ **`VectorStore` ê°ì²´**ëŠ” [**Runnable**](https://python.langchain.com/api_reference/core/index.html#langchain-core-runnables)ì˜ ì„œë¸Œí´ë˜ìŠ¤ê°€ ì•„ë‹™ë‹ˆë‹¤. ë°˜ë©´, **[Retrievers](https://python.langchain.com/api_reference/core/index.html#langchain-core-retrievers)** ëŠ” **Runnable**ì˜ ì„œë¸Œí´ë˜ìŠ¤ì…ë‹ˆë‹¤. ë”°ë¼ì„œ **ë™ê¸°(Synchronous)** ë° **ë¹„ë™ê¸°(Asynchronous)** `invoke`ì™€ `batch` ì‘ì—…ê³¼ ê°™ì€ í‘œì¤€ ë©”ì„œë“œ ì§‘í•©ì„ êµ¬í˜„í•©ë‹ˆë‹¤.  

### **ê²€ìƒ‰ê¸° vs ë²¡í„° ìŠ¤í† ì–´**  
- **ë²¡í„° ìŠ¤í† ì–´(VectorStore)**: ë²¡í„°í™”ëœ ë°ì´í„°ë¥¼ ì €ì¥í•˜ê³  ì¿¼ë¦¬í•©ë‹ˆë‹¤.  
- **ê²€ìƒ‰ê¸°(Retriever)**: ë²¡í„° ìŠ¤í† ì–´ë¥¼ í™œìš©í•˜ê±°ë‚˜ ì™¸ë¶€ APIì™€ ê°™ì€ **ë¹„ë²¡í„° ë°ì´í„° ì†ŒìŠ¤**ì™€ë„ í†µì‹ í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.  

---

## **ê²€ìƒ‰ê¸° ì§ì ‘ ìƒì„±í•˜ê¸°**  

`Retriever`ë¥¼ ì„œë¸Œí´ë˜ì‹±í•˜ì§€ ì•Šê³ ë„ ê°„ë‹¨í•œ ë²„ì „ì„ ì§ì ‘ ìƒì„±í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.  
- ì›í•˜ëŠ” **ë¬¸ì„œ ê²€ìƒ‰ ë©”ì„œë“œ**ë¥¼ ì„ íƒí•©ë‹ˆë‹¤.  
- í•´ë‹¹ ë©”ì„œë“œë¥¼ ì‚¬ìš©í•´ **Runnable** ê°ì²´ë¥¼ ì‰½ê²Œ ìƒì„±í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.  

###  **ì˜ˆì‹œ: `similarity_search`ë¥¼ í™œìš©í•œ ê²€ìƒ‰ê¸°**  

ë‹¤ìŒ ë‹¨ê³„ì—ì„œëŠ” `similarity_search` ë©”ì„œë“œë¥¼ ê¸°ë°˜ìœ¼ë¡œ ê°„ë‹¨í•œ ê²€ìƒ‰ê¸°ë¥¼ êµ¬ì¶•í•´ ë³´ê² ìŠµë‹ˆë‹¤.

````python
from typing import List  
from langchain_core.documents import Document
from langchain_core.runnables import chain

# ì‚¬ìš©ì ì •ì˜ ê²€ìƒ‰ê¸°(Retriever) ìƒì„±
# ì´ ê²€ìƒ‰ê¸°ëŠ” ì£¼ì–´ì§„ ì¿¼ë¦¬(query)ë¥¼ ê¸°ë°˜ìœ¼ë¡œ ë²¡í„° ìŠ¤í† ì–´ì—ì„œ ë¬¸ì„œë¥¼ ê²€ìƒ‰í•©ë‹ˆë‹¤.
@chain
def retriever(query: str) -> List[Document]:
    return vector_store.similarity_search(query, k=1)  # k=1ë¡œ ê°€ì¥ ìœ ì‚¬í•œ ë¬¸ì„œ 1ê°œ ë°˜í™˜

# ì—¬ëŸ¬ ê°œì˜ ì¿¼ë¦¬ë¥¼ ë™ì‹œì— ë°°ì¹˜(batch) ê²€ìƒ‰ ì‹¤í–‰í•˜ì—¬ ê²€ìƒ‰ ê²°ê³¼ë¥¼ ë°˜í™˜
retriever.batch(
    [
        "2023ë…„ Nikeì˜ ë§¤ì¶œì€ ì–¼ë§ˆì˜€ë‚˜ìš”?",  # ì²« ë²ˆì§¸ ì¿¼ë¦¬
        "2023ë…„ Nikeì˜ ë§ˆì§„ì€ ì–´ë–»ê²Œ ì˜í–¥ì„ ë°›ì•˜ë‚˜ìš”?",  # ë‘ ë²ˆì§¸ ì¿¼ë¦¬
    ],
)
````

**ë²¡í„° ìŠ¤í† ì–´(Vectorstores)** ëŠ” **`as_retriever`** ë©”ì„œë“œë¥¼ êµ¬í˜„í•˜ë©°, ì´ë¥¼ í†µí•´ **[VectorStoreRetriever](https://python.langchain.com/api_reference/core/vectorstores/langchain_core.vectorstores.base.VectorStoreRetriever.html)** ë¥¼ ìƒì„±í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.  

ì´ ê²€ìƒ‰ê¸°(Retriever)ëŠ” íŠ¹ì • **`search_type`** ë° **`search_kwargs`** ì†ì„±ì„ í¬í•¨í•˜ë©°, ì´ë¥¼ í†µí•´ **ë²¡í„° ìŠ¤í† ì–´ì˜ ì–´ë–¤ ë©”ì„œë“œë¥¼ í˜¸ì¶œí• ì§€**, ê·¸ë¦¬ê³  **ì–´ë–»ê²Œ ë§¤ê°œë³€ìˆ˜ë¥¼ ì„¤ì •í• ì§€**ë¥¼ ì •ì˜í•©ë‹ˆë‹¤.  

ì˜ˆë¥¼ ë“¤ì–´, ìœ„ì˜ ì˜ˆì œë¥¼ ë‹¤ìŒê³¼ ê°™ì´ ì¬í˜„í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤:  

````python
# `as_retriever` ë©”ì„œë“œë¥¼ ì‚¬ìš©í•˜ì—¬ ë²¡í„° ìŠ¤í† ì–´ë¥¼ ê²€ìƒ‰ê¸° ê°ì²´ë¡œ ë³€í™˜í•©ë‹ˆë‹¤.

retriever = vector_store.as_retriever(
    search_type="similarity",  # ê²€ìƒ‰ ìœ í˜•ì„ 'similarity'(ìœ ì‚¬ë„)ë¡œ ì„¤ì •
    search_kwargs={"k": 1},  # ìƒìœ„ 1ê°œì˜ ìœ ì‚¬í•œ ê²°ê³¼ë§Œ ë°˜í™˜í•˜ë„ë¡ ì„¤ì •
)

# ì—¬ëŸ¬ ê°œì˜ ì¿¼ë¦¬ë¥¼ ë™ì‹œì— ì‹¤í–‰í•˜ì—¬ ê²€ìƒ‰ ê²°ê³¼ë¥¼ ë°˜í™˜í•©ë‹ˆë‹¤.
retriever.batch(
    [
        "2023ë…„ Nikeì˜ ë§¤ì¶œì€ ì–¼ë§ˆì˜€ë‚˜ìš”?",  # ì²« ë²ˆì§¸ ì¿¼ë¦¬: 2023ë…„ Nikeì˜ ë§¤ì¶œ ì •ë³´
        "2023ë…„ Nikeì˜ ë§ˆì§„ì€ ì–´ë–»ê²Œ ì˜í–¥ì„ ë°›ì•˜ë‚˜ìš”?",  # ë‘ ë²ˆì§¸ ì¿¼ë¦¬: 2023ë…„ Nikeì˜ ë§ˆì§„ ì˜í–¥
    ],
)
````

## êµ¬í˜„ëœ ì‹œë§¨í‹± ê²€ìƒ‰ ì—”ì§„ì„ ì´ìš©í•œ RAG Example

````python
from langchain_core.prompts import PromptTemplate
from langchain_openai import ChatOpenAI
from langchain.chains import RetrievalQA

# 1ï¸ê²€ìƒ‰ê¸°(Retriever) ì„¤ì •
retriever = vector_store.as_retriever(search_type="similarity", search_kwargs={"k": 3})

# RAG í”„ë¡¬í”„íŠ¸ ì„¤ì •
rag_prompt = PromptTemplate(
    input_variables=["context", "question"],
    template="""
    ì£¼ì–´ì§„ ì»¨í…ìŠ¤íŠ¸ë¥¼ ê¸°ë°˜ìœ¼ë¡œ ì‚¬ìš©ìì˜ ì§ˆë¬¸ì— ë‹µí•˜ì„¸ìš”.
    
    ì»¨í…ìŠ¤íŠ¸:
    {context}
    
    ì§ˆë¬¸:
    {question}
    """,
)

# LLM ì„¤ì •
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)

# RAG íŒŒì´í”„ë¼ì¸ (ê²€ìƒ‰ + ë‹µë³€ ìƒì„±)
rag_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",  # ë¬¸ì„œë¥¼ í•œ ë²ˆì— ë„£ì–´ ì²˜ë¦¬
    retriever=retriever,
    chain_type_kwargs={"prompt": rag_prompt}
)

# 5ï¸âƒ£ ì§ˆë¬¸ ì‹¤í–‰
query = "2023ë…„ Nikeì˜ ë§¤ì¶œì€ ì–¼ë§ˆì˜€ë‚˜ìš”?"
response = rag_chain.invoke(query)

# 6ï¸âƒ£ ê²°ê³¼ ì¶œë ¥
print("ğŸ”¹ ì§ˆë¬¸:", query)
print("ğŸ”¹ ì‘ë‹µ:", response)
````
